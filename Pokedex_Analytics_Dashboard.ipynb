```python
import requests
import pandas as pd

# 1. Buscando dados do Pikachu
pokemon = "pikachu"
url_poke = f"https://pokeapi.co/api/v2/pokemon/{pokemon}"
resposta_poke = requests.get(url_poke)
dados_poke = resposta_poke.json()

# 2. Navegando no JSON complexo (Nível Sênior)
estatisticas = []
for stat in dados_poke['stats']:
    nome_stat = stat['stat']['name']
    valor_stat = stat['base_stat']
    estatisticas.append({'Estatística': nome_stat, 'Valor Base': valor_stat})

# 3. Visualização
df_stats = pd.DataFrame(estatisticas)
print(f"--- Status Base do {pokemon.capitalize()} ---")
display(df_stats)
```

```python
# Célula de Execução: Top 10 Pokémon
import requests
import pandas as pd
import time

# 1. Lista vazia para armazenar os dicionários com dados
dados_pokemon = []

print("Buscando dados na PokeAPI. Aguarde...")

# 2. Loop para buscar os IDs de 1 a 10
for i in range(1, 11):
    url = f"https://pokeapi.co/api/v2/pokemon/{i}"
    resposta = requests.get(url)

    # Tratamento básico de erro (valida o status code)
    if resposta.status_code == 200:
        dados = resposta.json()

        # Extraindo dados do JSON complexo
        nome = dados['name'].capitalize()
        exp_base = dados['base_experience']
        peso_kg = dados['weight'] / 10

        # Adicionando à nossa lista
        dados_pokemon.append({
            'ID': i,
            'Nome': nome,
            'Experiência Base': exp_base,
            'Peso (kg)': peso_kg
        })

    # BOAS PRÁTICAS: Pausa para não sobrecarregar a API
    time.sleep(0.5)

# 3. Transformando a lista em uma tabela organizada (DataFrame)
df_poke = pd.DataFrame(dados_pokemon)

# 4. Ordenando para ver o "Top 10" com maior experiência
df_top10 = df_poke.sort_values(by="Experiência Base", ascending=False).reset_index(drop=True)

# 5. Visualização no Colab
print("\n--- Top Pokémon por Experiência Base ---")
display(df_top10)
```

```python
import pandas as pd
import requests
import plotly.express as px
import time

# 1. Buscando os dados (Top 20 para um ranking legal)
dados_pokemon = []
print("Buscando dados na PokeAPI... Aguarde.")

for i in range(1, 21):
    url = f"https://pokeapi.co/api/v2/pokemon/{i}"
    resposta = requests.get(url)

    if resposta.status_code == 200:
        dados = resposta.json()
        dados_pokemon.append({
            'Nome': dados['name'].capitalize(),
            'Experiência Base': dados['base_experience']
        })
    time.sleep(0.2) # Pausa segura para não sobrecarregar a API

# 2. Processamento dos Dados
df_poke = pd.DataFrame(dados_pokemon)
df_ranqueado = df_poke.sort_values(by="Experiência Base", ascending=True)

# 3. Criando o Gráfico de Barras com Plotly
fig = px.bar(
    df_ranqueado,
    x="Experiência Base",
    y="Nome",
    orientation='h',
    color="Experiência Base",
    color_continuous_scale="Viridis",
    title="🏆 Ranking de Pokémon por XP Base"
)

# 4. Mostrando o gráfico na tela do Colab
fig.show()
```

```python
%%writefile app.py
import streamlit as st
import pandas as pd
import requests
import plotly.express as px

st.set_page_config(page_title="Pokedex Analytics", layout="centered")
st.title("🏆 Ranking de Experiência Pokémon")
st.write("Criado por um futuro Analista Sênior!")

@st.cache_data
def buscar_dados_pokemon(limite=20):
    dados_pokemon = []
    for i in range(1, limite + 1):
        url = f"https://pokeapi.co/api/v2/pokemon/{i}"
        resposta = requests.get(url)
        if resposta.status_code == 200:
            dados = resposta.json()
            dados_pokemon.append({
                'Nome': dados['name'].capitalize(),
                'Experiência Base': dados['base_experience']
            })
    return pd.DataFrame(dados_pokemon)

with st.spinner("Buscando dados na PokeAPI..."):
    df_poke = buscar_dados_pokemon(20)

# Processando os dados
df_ranqueado = df_poke.sort_values(by="Experiência Base", ascending=True)

# Gerando o gráfico
fig = px.bar(
    df_ranqueado, x="Experiência Base", y="Nome", orientation='h',
    color="Experiência Base", color_continuous_scale="Viridis"
)

st.plotly_chart(fig, use_container_width=True)
```
